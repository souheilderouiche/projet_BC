```python
####Fonction Voisins Communs
```

```python
def Voisins_Communs(g,x,y):
    Nx=g.neighbors(x)
    Ny=g.neighbors(y)
    return len(list(set(Nx) & set(Ny)))
Voisins_Communs(g,1,4)
```

```python
# Affichage des 5 premiers sommets, tri de mani√®re croissante
valeurs = {'label':g.vs['label'],'degree':g.vs.degree()}
print(pandas.DataFrame.from_dict(valeurs).sort_values(by='degree',ascending=False).iloc[:5,:])
```

```python
# Closeness centralit√©
valeurs = {'node_name':g.vs['label'],'closeness':g.vs.closeness()}
print(pandas.DataFrame.from_dict(valeurs).sort_values(by='closeness',ascending=False).iloc[:5,:])
```
